```python
#6
n=int(input("Enter a number lesser than 10:"))
for i in range(1,n+1):
    for j in range(i,n+1):
        if i==1:
            print(j,end=" ")
        elif i==j:
            print((str(i)+" ")*i,end="")
        else:
            print(j,end=" ")
    print()
```

```python
#13
#Upper Triangle
n=int(input("Enter number:"))
print()
x=1
for i in range(n,0,-1):
    print(x*" "+"* "*i)
    x+=1
#Lower Triangle
x=n
for i in range(1,n+1):
    print(x*" "+"* "*i)
    x-=1
```
